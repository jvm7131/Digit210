```python
import spacy
from collections import Counter
import pygal
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
filepath = "s1/01x11_-_Stella.txt"

with open(filepath, "r", encoding="utf-8") as f:
    text = f.read()
```

```python
doc = nlp(text)
```

```python
verbs = []

for token in doc:
    if token.pos_ == "VERB":
        verbs.append(token.lemma_)
```

```python
word_freq = Counter(verbs)
ranked = word_freq.most_common()


topTen = word_freq.most_common(10)
print(topTen)
```

```python
with open("verbFreq.txt", "w", encoding="utf-8") as o:
    for w, num in ranked:
        o.write(str(w) + " :::: " + str(num) + "\n")
```

```python
bar_chart = pygal.Bar()
bar_chart.title = "Top 10 Verbs in Stella Episode"

for word, freq in topTen:
    bar_chart.add(word, freq)

bar_chart.render_to_file("top10_verbs.svg")
```
