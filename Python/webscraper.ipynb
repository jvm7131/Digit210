```python
import bs4
import requests
import os


archive_url = "http://www.textfiles.com/humor/"

def get_linkContents():
    r = requests.get(archive_url)
    print(r)  

    soup = bs4.BeautifulSoup(r.content, 'html.parser')

    linkText = []

    
    for item in soup.find_all('a'):
        href = item.get('href')
        text = item.text.strip()

        if href and href.endswith('.txt'):
            print(text)
            linkText.append(text)

    download_toFile(linkText)

def download_toFile(linkText):
    file_name = "outputLinkText.txt"
    working_dir = os.getcwd()
    file_deposit = os.path.join(working_dir, file_name)

    with open(file_deposit, 'w', encoding='utf-8') as f:
        for item in linkText:
            f.write(item + '\n')

    print("Saved file to:", file_deposit)


get_linkContents()
```
