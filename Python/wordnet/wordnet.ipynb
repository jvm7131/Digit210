```python
import os
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.corpus import wordnet as wn

nltk.download("punkt")
nltk.download("averaged_perceptron_tagger")
nltk.download("wordnet")
nltk.download("omw-1.4")
```

```python
cwd = os.getcwd()
print("Current working directory:")
print(cwd)
```

```python
filepath = "s1/01x01_-_Operation_Strix.txt"

text = open(filepath, "r", encoding="utf-8").read()

print(text[:500])
```

```python
tokens = word_tokenize(text)

print("Number of tokens:", len(tokens))
print(tokens[:50])
```

```python
lowercase_tokens = [token.lower() for token in tokens]

unique_tokens = set(lowercase_tokens)

print("Number of unique tokens:", len(unique_tokens))
print(list(unique_tokens)[:50])
```

```python
tagged_tokens = pos_tag(tokens)

print(tagged_tokens[:50])
```

```python
adjectives = [word.lower() for word, tag in tagged_tokens if tag == "JJ"]

unique_adjectives = sorted(set(adjectives))

print("Number of adjective tokens:", len(adjectives))
print("Number of unique adjectives:", len(unique_adjectives))
print(unique_adjectives[:50])
```

```python
adjective_results = []

for word in unique_adjectives:
    lemma = wn.morphy(word)
    
    if lemma is None:
        lemma = word
    
    synsets = wn.synsets(lemma)
    synset_count = len(synsets)
    
    adjective_results.append((word, lemma, synset_count))

print(adjective_results[:30])
```

```python
sorted_results = sorted(adjective_results, key=lambda item: item[2], reverse=True)

print("Most ambiguous adjectives:")
for word, lemma, count in sorted_results[:20]:
    print(f"{word} | lemma: {lemma} | synsets: {count}")
```

```python
os.makedirs("outputs", exist_ok=True)

print("Output folder is ready.")
```

```python
output_path = "outputs/adjective_wordnet_results.txt"

with open(output_path, "w", encoding="utf-8") as outfile:
    outfile.write("word\tlemma\tsynset_count\n")
    
    for word, lemma, count in sorted_results:
        outfile.write(f"{word}\t{lemma}\t{count}\n")

print(f"Results saved to {output_path}")
```

```python
top_output_path = "outputs/top_20_ambiguous_adjectives.txt"

with open(top_output_path, "w", encoding="utf-8") as outfile:
    outfile.write("Top 20 Most Ambiguous Adjectives\n\n")
    
    for word, lemma, count in sorted_results[:20]:
        outfile.write(f"{word} | lemma: {lemma} | synsets: {count}\n")

print(f"Top 20 results saved to {top_output_path}")
```
